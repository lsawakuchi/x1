```python
import pandas as pd
import numpy as np
from datetime import datetime
from conector.mysql import mysql_engine, CaptalysDBContext
from dateutil.relativedelta import relativedelta
import matplotlib.pyplot as plt
import json
from datamanager import conn_pricing


class AssetManager(object):
    def __init__(self, df_acompanhamento=None):
        self.acompanhamento = self.getAcompanhamento() if df_acompanhamento is None else df_acompanhamento
    
    @staticmethod
    def getAcompanhamento():
        query = "select * from acompanhamento"
        session = conn_pricing.new_session()
        res = session.execute(query).fetchall()
        session.close()
        acompanhamento = [dict(el) for el in res]
        df = pd.DataFrame(acompanhamento)
        df['valorAquisicao'] = df["valorAquisicao"].astype(float)

        df['status'] = df.apply(lambda x: 'CRITICO' if x['status'] in ['CRITICO, NENHUM PAGAMENTO REALIZADO',
                                "NENHUM PAGAMENTO REALIZADO"] else
                                ("QUITADA" if x['status'] == "QUITADA a VP" else x["status"]), axis=1)
        df['produto'] = df.apply(lambda x: x["produto"].upper(), axis=1)
        

        return df
    
    @staticmethod
    def formataValor(val):
        val = str(val).replace(".", ",")
        lval = val.split(",")
        v1 = lval[0]
        v2 = lval[1]
        if len(v1) <= 3:
            valor_final = v1
        else:
            valor_final = v1[:len(v1) % 3]
            p0 = len(v1) % 3
            for i in range(int((len(v1) - len(v1) % 3)/3)):
                valor_final = valor_final + "." + v1[p0:p0+3]
                p0 = p0+3
        if len(v2) == 1:
            v2 = v2 + '0'
        if valor_final[0] == '.':
            valor_final = valor_final[1:]
        valor_final = "R$ " + valor_final + "," + v2
        return valor_final
    
    @property
    def colorMenu(self):
        return {'ALERTA': '#FCF663',
                'QUITADA': '#63E89E',
                'BOM': '#A2F2F5',
                'CRITICO': '#FB6868',
                'OTIMO': '#0B81B9',
                'SEM PAGAMENTO': '#F9930F',
                'RENEGOCIADO': '#938A8A',
                'EXECUCAO' : '#FB1D46',
                'REESTRUTURADO' : '#983975',
                'COBRANCA' : '#8B94F1',
                'CRITICO' :'#F9D511',
                "status": "#63E89E",
                'produto': "#57989C",
                "tomatico": "#F88D8D",
                "wirecard": "#56A1C8",
                "creditoveloz": "#5ECD82",
                }
    
    @property
    def status_(self):
        return {"regular" : ["QUITADA", "BOM", "OTIMO", "ALERTA"],
                "irregular" : ["EXECUCAO", " REESTRUTURADO", "COBRANCA", "CRITICO"],
                "geral" : ["QUITADA", "BOM", "OTIMO", "ALERTA", "EXECUCAO", " REESTRUTURADO", "COBRANCA", "CRITICO"]
               }
        


    def plotStatus(self, status="regular"):
        df = self.acompanhamento[["cnpj", "produto", "status"]]
#         df["status"] = df.apply(lambda x: "QUITADA" if x["status"] == "QUITADA a VP" else x["status"], axis=1)
#         df['status'] = df.apply(lambda x: "SEM PAGAMENTO" if x['status'] in ['CRITICO, NENHUM PAGAMENTO REALIZADO', 'NENHUM PAGAMENTO REALIZADO'] else x['status'], axis=1)
        lista_status = self.status_.get(status)
        df = df[df['status'].isin(lista_status)]
        res = df.groupby(["produto", "status"]).count().reset_index()

        i = 1
        fig, ax = plt.subplots(figsize=(12, 7))
        lista_produto = res['produto'].unique().tolist()
        if len(lista_produto)==3:
            for el in lista_produto:
                dt = res[res['produto'] == el]
                dt['prop'] = dt['cnpj']/dt['cnpj'].sum()
                sizes = dt['prop'].tolist()
                sizes = [el*100 for el in sizes]
                labels = dt['status'].tolist()
                colors = [self.colorMenu.get(el) for el in labels]

                plt.subplot(1, 3, i)

                plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
                        shadow=True, startangle=90)
                i += 1

                plt.axis('equal')
                plt.title(el)
                plt.tight_layout()

    #         fig.savefig('../img/distribuicao_status_produto.png', bbox_inches="tight")
            fig.savefig('distribuicao_status_{}_produto.png'.format(status), bbox_inches="tight")
        
        if len(lista_produto)==4:
            for el in lista_produto:
                dt = res[res['produto'] == el]
                dt['prop'] = dt['cnpj']/dt['cnpj'].sum()
                sizes = dt['prop'].tolist()
                sizes = [el*100 for el in sizes]
                labels = dt['status'].tolist()
                colors = [self.colorMenu.get(el) for el in labels]

                plt.subplot(2, 2, i)

                plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
                        shadow=True, startangle=90)
                i += 1

                plt.axis('equal')
                plt.title(el)
                plt.tight_layout()

    #         fig.savefig('../img/distribuicao_status_produto.png', bbox_inches="tight")
            fig.savefig('distribuicao_status_{}_produto.png'.format(status), bbox_inches="tight")
        return

    @staticmethod
    def getTaxa():

        query = "select * from taxa_carteira"
        session = conn_pricing.new_session()
        res = session.execute(query).fetchall()
        session.close()

        res = [dict(el) for el in res]
        df_taxa = pd.DataFrame(res)
        df_taxa["data"] = df_taxa.apply(lambda x: x['dataCalculo'].date(), axis=1)
        df_taxa = df_taxa[df_taxa['data'] == df_taxa['data'].max()]

        df_taxa["status"] = df_taxa.apply(lambda x: "CRITICO" if x['status'] == "CRITICO, NENHUM PAGAMENTO REALIZADO" else x["status"], axis=1)
        df_taxa["status"] = df_taxa.apply(lambda x: "CARTEIRA" if x["status"] == "TODOS" else x["status"], axis=1)
        df_taxa = df_taxa[["produto", "status", "taxa"]]
        return df_taxa

    def plotTaxa(self):
        df_taxa = self.getTaxa()

        filtro = list(df_taxa.columns)[0]
        fig, ax = plt.subplots(figsize=(12, 7))
        opacity = 0.4
        error_config = {'ecolor': '0.3'}

        x = df_taxa.iloc[:, 0].tolist()
        print(x)
        y = df_taxa["taxa"].tolist()

        ax.bar(x, y, alpha = opacity, color = self.colorMenu.get(filtro), error_kw=error_config, edgecolor="black")
        plt.ylabel("taxa mensal (%)", fontsize=13)
#         plt.savefig('../img/taxa_carteira_{}.png'.format(filtro))
        plt.savefig('../img/taxa_carteira_{}.png'.format(filtro))
        
        return

    def plotDesembolso(self):
        df_des = self.acompanhamento.groupby("status").sum().reset_index()[["status", "valorAquisicao"]]
        df_des.columns = ['status', 'desembolso']
#         filtro = list(df_taxa.columns)[0]
        fig, ax = plt.subplots(figsize=(12, 7))
        opacity = 0.8
        error_config = {'ecolor': '0.3'}

        x = df_des["status"].tolist()
        y = df_des["desembolso"].tolist()

        ax.bar(x, y, alpha = opacity, color = "#7FCAB0", error_kw=error_config, edgecolor="black")
        plt.ylabel("desembolso (R$)", fontsize=13)
#         plt.savefig('../img/desembolso_status.png')
        plt.savefig('desembolso_status.png')
        
        return
    
    def getDesembolsos(self, periodo=7):
        data_inicio = datetime.now().date() - relativedelta(days=periodo)
        dfnew = self.acompanhamento[self.acompanhamento['dataCessao'] > data_inicio]
        dfop = dfnew.groupby("produto").count().reset_index()[["produto", "cnpj"]]
        dfop.columns = ['produto', "quantidade"]
        dfval = dfnew.groupby("produto").sum().reset_index()[["produto", "valorAquisicao"]]
        dfval.columns = ['produto', 'desembolso']
        dfop = dfop.merge(dfval, left_on='produto', right_on='produto')
        return dfop.to_dict("records")

    def analise_ml(self):
        df_acompanhamento = self.acompanhamento
        df_acompanhamento["retencao_ajustada"] = df_acompanhamento.apply(lambda x: json.loads(x["milestones"]).get("tx_retencao_ajustada") if not x["milestones"] is None else 0, axis=1)
        dfml = df_acompanhamento[df_acompanhamento['status'] != 'QUITADA']
        # wirecard nao usa milestones
        dfml = dfml[dfml['produto'] != 'WIRECARD']
        dfml['flag_ml'] = dfml.apply(lambda x: int(len(json.loads(x['milestones'])) > 0) if not x["milestones"] is None else 0, axis=1)
        dfml_ = dfml[dfml['flag_ml'] == 1][["cnpj", 'empresa', 'produto', 'controleParticipante', 'retencao_ajustada', 'dataCessao']]
        dfml_.fillna(0, inplace=True)
        now = datetime.now().date()
        dfml_["tempo_operacao"] = dfml_.apply(lambda x: (now - x['dataCessao']).days, axis=1)
        return dfml_

    @staticmethod
    def plotMilestones(df_milestones):
        x = df_milestones[(df_milestones['retencao_ajustada'] > 0) & (df_milestones['retencao_ajustada'] < 1)]["retencao_ajustada"].tolist()
        x = [el*100 for el in x]
        fig, ax = plt.subplots(figsize=(12, 7))
        plt.hist(x, bins=8, color='#A494B3')
        plt.ylabel("quantidade de operacoes", fontsize=13)
        plt.xlabel("retencao ajustada (%)", fontsize=13)
        plt.savefig('../img/ajuste_retencao.png')
        return

    def plotEvolucaoPagamentos(self):
        dt = self.acompanhamento[["cnpj", 'controleParticipante', 'valorDevido', 'valorPago']]

        dt["evolucao"] = dt["valorPago"]/dt["valorDevido"]

        dt['evolucao'] = dt['evolucao'].astype(float)

        dt["evolucao"] = dt.apply(lambda x: np.around(100*x['evolucao'], 2), axis=1)

        x1 = dt[dt["evolucao"] <= 20]["evolucao"].tolist()
        x2 = dt[(dt["evolucao"] >= 20) & (dt["evolucao"] <= 40)]["evolucao"].tolist()
        x3 = dt[(dt["evolucao"] >= 40) & (dt["evolucao"] <= 60)]["evolucao"].tolist()
        x4 = dt[(dt["evolucao"] >= 60) & (dt["evolucao"] <= 80)]["evolucao"].tolist()
        x5 = dt[dt["evolucao"] >= 80]["evolucao"].tolist()

        fig, ax = plt.subplots(figsize=(12, 7))
        plt.hist(x1, color='#2DC4D6', bins=5)
        plt.hist(x2, color='#36A5B5', bins=5)
        plt.hist(x3, color='#2C8BB8', bins=4)
        plt.hist(x4, color='#1A5F7F', bins=3)
        plt.hist(x5, color='#115161', bins=2)
        plt.ylabel("quantidade de operações", fontsize=13)
        plt.xlabel('percentual de pagamento (%)', fontsize=13)
        plt.savefig("../img/evolucao_pagamento.png")
        return


# if __name__ == '__main__':
#     am = AssetManager()
#     am.plotDesembolso()
#     print(am.getAcompanhamento())
```
