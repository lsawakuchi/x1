- I will aproach the task of detecting outlier behavior related to debt quality by applying **Isolation Forest**
- Next, I will apply **PCA** to identify the outliers graphically

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("../tabelas/dataset_modelo.xlsx")
```

```python
df.head()
```

####  Outlier, once applied elegibility conditions

- Applying elegibility conditions

```python
dataset = df[(df['prop_divida']<1.5) & (df["quantidade_cheques"]==0)]
# dataset = df.copy()
```

```python
# dataset.drop(columns=['prop_divida', 'quantidade_cheques', 'outros'], axis=1, inplace=True)
dataset.drop(columns=['outros', 'prop_divida', 'outros', 'quantidade_cheques'], axis=1, inplace=True)
```

```python
dataset.max()
```

```python
dataset.head()
```

```python
dataset = dataset[['cnpj', 'credito', 'infra', 'processos', 'dispersao']]
```

```python
dataset.index = dataset.cnpj
```

```python
dataset.drop(columns=['cnpj'], axis=1, inplace=True)
```

```python
dataset.head()
```

- The features **idade_maxima** and **tempo_medio** may determine an outlier, but it depends on the values of the features to decide whether or not the outlier should be red flagged

```python
# feature dataframe
# X = dataset.iloc[:, 1:]
X = dataset.copy()
```

```python
X.head()
```

```python
X.shape
```

### Isolation Forest

```python
outlier_detect =  IsolationForest(n_estimators=1000,
                                max_samples=100, 
                                contamination=0.05, 
                                max_features=X.shape[1], 
                                random_state=1)
```

```python
outlier_detect.fit(X)
```

```python
outliers_predicted = outlier_detect.predict(X)
```

```python
dataset["outlier"] = outliers_predicted
```

```python
df_out = dataset[dataset['outlier']==-1]
```

### PCA

```python
features = list(X.columns)
x = X.loc[:, features].values
x = StandardScaler().fit_transform(x)
X = pd.DataFrame(x)
X.columns = features
```

```python
X.head()
```

```python
pca = PCA(n_components=2)

principalComponents = pca.fit_transform(X)

df_pca = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])

df_pca.head()
```

```python
df_pca['outlier'] = outliers_predicted
```

```python
df_pca.head()
```

```python
df_pca.shape
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ["outlier", "inlier"]
colors = ['r', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df_pca['outlier'] == -1 if target == "outlier" else df_pca['outlier'] == 1
    ax.scatter(df_pca.loc[indicesToKeep, 'pc1']
               , df_pca.loc[indicesToKeep, 'pc2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
df_pca.index = dataset.index
```
