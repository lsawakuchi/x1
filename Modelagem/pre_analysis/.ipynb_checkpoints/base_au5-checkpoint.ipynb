```python
df["data"] = df.apply(lambda x : x["data"] - relativedelta(months=1), axis=1)
```

```python
fr = []
for el in df['cnpj'].unique().tolist():
    body = {"dados" : df[df['cnpj']==el][["data", "valor"]].to_dict('records'), "id_produto" : 'tomatico'}
    ls = LScoring(body)
    res = ls.calcula()
    fr.append(pd.DataFrame({'cnpj' : [el], 'score' : [res.get('score')]}))
```
