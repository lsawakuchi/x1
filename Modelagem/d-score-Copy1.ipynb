```python
def probability(dt):
    dprob = dt.groupby("segmento").count().reset_index()[["segmento", "cnpj"]]
    dprob.columns = ["segmento", "ocorrencias"]
    dprob["probabilidade"] = dprob["ocorrencias"]/dprob["ocorrencias"].sum()
    return dprob
```

```python
probability(dt)
```

```python
def impact(dt):
    dimp = dt.groupby(["segmento"]).sum().reset_index()
    dimp["impact"] = dimp["valor"]/dimp["valor"].sum()
    dimp["impact"] = dimp.apply(lambda x : x["impact"]*I.get(x["segmento"]), axis=1)
    return dimp
```

```python
impact(dt)
```

```python
def get_risk(cnpj, df):
    dt = df[df['cnpj']==cnpj]
    dt.drop_duplicates(inplace=True)
    dprob = probability(dt)
    dimp = impact(dt)
    
    drisk = dprob.merge(dimp, left_on="segmento", right_on="segmento", how="left")
    drisk["risk"] = drisk["probabilidade"]*drisk["impact"]
    drisk["weight"] = drisk.apply(lambda x : W.get(x["segmento"]), axis=1)
    drisk["total_risk"] = drisk["risk"]*drisk["weight"]
    risk = drisk["total_risk"].sum()/10
    return risk
```
