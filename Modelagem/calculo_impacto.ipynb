```python
df['valor'] = df.apply(lambda x : x['valor'].split("R$ ")[1].replace(",", "."), axis=1)

df["valor"] = df.apply(lambda x : float(x["valor"]), axis=1)

df["valor"].iloc[0]
```

```python
df_comp = pd.read_excel("../tabelas/dataset_variaveis_completo_201904.xlsx")

df_comp["cnpj"] = df_comp.apply(lambda x : str(x["cnpj"]), axis=1)

df_comp["cnpj"] = df_comp.apply(lambda x : "0" + x["cnpj"] if len(x["cnpj"])==13 else
                                ("00" + x["cnpj"] if len(x["cnpj"])==12 else x["cnpj"]), axis=1)
```

```python
def probabilidade(dt):
    dprob = dt.groupby("segmento").count().reset_index()[["segmento", "cnpj"]]
    dprob.columns = ["segmento", "ocorrencias"]
    dprob["probabilidade"] = dprob["ocorrencias"]/dprob["ocorrencias"].sum()
    return dprob

def impacto(dt):
    dt["peso"] = dt.apply(lambda x : I.get(x["segmento"]), axis=1)
    dt["valor_impacto"] = dt["valor"]*dt["peso"]
    dimp = dt.groupby(["segmento"]).sum().reset_index()[["segmento", "valor_impacto", "valor"]]
    dimp["impacto_pond"] = dimp["valor_impacto"]/dimp["valor_impacto"].sum()
    dimp["impacto"] = dimp["valor"]/dimp["valor"].sum()
    return dimp

I = {"credito" : 4, "processos" : 3, "infra" : 2, "outros" : 1}

def calcula_risco(cnpj, df_divida):
    dt = df[df["cnpj"]==cnpj]
    dprob = probabilidade(dt)
    dimp = impacto(dt)
    drisco = dprob.merge(dimp, left_on="segmento", right_on="segmento", how="left")
    drisco["risco"] = drisco["impacto"]*drisco["probabilidade"]
    drisco["risco_pond"] = drisco["impacto_pond"]*drisco["probabilidade"]
    drisco["cnpj"] = cnpj
    return drisco
```

```python
fr = []
for el in df["cnpj"].unique().tolist():
    dr = calcula_risco(el, df)
    fr.append(dr)
```

```python
escala_impacto = {"outros" : {"c0" : 0, "c1" : 0.25},
                 "infra" : {"c0" : 0.25, "c1" : 0.5},
                 "processos" : {"c0" : 0.5 , "c1" : 0.75},
                 "credito" : {"c0" : 0.75, "c1" : 1}
                 }

def get_impacto_segmento(comp, segmento, escala):
    escala_seg = escala.get(segmento)
    c0 = escala_seg.get("c0")
    c1 = escala_seg.get("c1")
    return (c1-c0)*comp + c0
```

```python
escala_score = {"credito" : {"s0" : 0, "s1" : 200},
               "processos" : {"s0" : 200, "s1" : 400},
                "infra" : {"s0" : 400, "s1" : 600},
                "outros" : {"s0" : 600, "s1" : 800}
               }

def dscore(r, seg, escala_score):
    escala = escala_score.get(seg)
    s0 = escala.get("s0")
    s1 = escala.get("s1")
    return (s0 - s1)*r + s1
```
