```python
def probability(dt):
    dprob = dt.groupby("segmento").count().reset_index()[["segmento", "cnpj"]]
    dprob.columns = ["segmento", "ocorrencias"]
    dprob["probabilidade"] = dprob["ocorrencias"]/dprob["ocorrencias"].sum()
    return dprob
```

```python
probability(dt)
```

```python
def impact(dt):
    dimp = dt.groupby(["segmento"]).sum().reset_index()
    dimp["impact"] = dimp["valor"]/dimp["valor"].sum()
    return dimp
```

```python
impact(dt)
```

```python
def get_risk(cnpj, df):
    dt = df[df['cnpj']==cnpj]
    dt.drop_duplicates(inplace=True)
    dprob = probability(dt)
    dimp = impact(dt)
    
    drisk = dprob.merge(dimp, left_on="segmento", right_on="segmento", how="left")
    drisk["risk"] = drisk["probabilidade"]*drisk["impact"]
    drisk["weight"] = drisk.apply(lambda x : W.get(x["segmento"]), axis=1)
    drisk["total_risk"] = drisk["risk"]*drisk["weight"]
    risk = drisk["total_risk"].sum()/10
    return risk
```

```python
resp = []
for el in df["cnpj"].unique().tolist():
    r = get_risk(el, df)
    resp.append(pd.DataFrame({"cnpj" : [el], "risk" : [r]}))
```

```python
resp = pd.concat(resp)
```

```python
resp["risk"].min()
```

```python
trace = go.Histogram(
    x = resp["risk"],
    marker = dict(color="#87EBB3", line=dict(color="rgb(0, 0, 0)", width=1))
)
layout = go.Layout(title="Distribution of Risk based on Debt quality", xaxis=dict(title="risk"),
                  yaxis=dict(title="number of samples"))
fig = go.Figure(data = [trace], layout=layout)
iplot(fig)
```

```python
# calculating the D-score
```

```python
def dscore(risk):
    return 1000*((-1)*(5/2)*risk + 1)
```

```python
dscore(0.2)
```

```python
resp["dscore"] = resp.apply(lambda x : dscore(x["risk"]), axis=1)
```

```python
resp.head()
```

```python
trace = go.Histogram(
    x = resp["dscore"],
    marker = dict(color="rgb(142,120,163)", line=dict(color="rgb(0, 0, 0)", width=1))
)
layout = go.Layout(title="Distribution of D-score - entire dataset")
fig = go.Figure(data = [trace], layout=layout)
iplot(fig)
```
