### 1 - Probabilidade

```python
def calcula_probabilidade(cnpj, df):
    dt = df[df["cnpj"]==cnpj]
    dt = dt.groupby("segmento").count().reset_index()[["segmento", "valor"]]
    dt.columns = ["segmento", "ocorrencias"]
    dt["probabilidade"] = dt["ocorrencias"]/dt["ocorrencias"].sum()
    dt["cnpj"] = cnpj
    return dt
```

```python
calcula_probabilidade(cnpj, df)
```

```python
ret = []
for el in df["cnpj"].unique().tolist():
    _df = calcula_probabilidade(el, df)
    ret.append(_df)
```

```python
dcalc = pd.concat(ret)

dcalc = dcalc[["cnpj", "segmento", "ocorrencias", "probabilidade"]]
```

### 2 -  Composicao da Dívida

```python
def calcula_composicao(cnpj, df):
    dt = df[df["cnpj"]==cnpj]
    dt = dt.groupby("segmento").sum().reset_index()
    dt.columns = ["segmento", "valor_divida"]
    dt["composicao"] = dt["valor_divida"]/dt["valor_divida"].sum()
    dt["cnpj"] = cnpj
    return dt
```

```python
ret = []
for el in df["cnpj"].unique().tolist():
    _df = calcula_composicao(el, df)
    ret.append(_df)
```

### Calculo do  $\lambda$ 

```python
dcalc["lambda"] = dcalc["composicao"]*dcalc["pi"]
```

```python
# escala do impacto
escala_impacto = {"credito" : {"i0" : 0.75, "i1" : 1},
                 "processos" : {"i0" : 0.5, "i1" : 0.75},
                  "infra" : {"i0" : 0.25, "i1" : 0.5},
                  "outros" : {"i0" : 0, "i1" : 0.25},
                 }
```

```python
escala_impacto
```

```python
def impacto_segmento(lambda_, segmento, escala_impacto):
    escala = escala_impacto.get(segmento)
    i0 = escala.get("i0")
    i1 = escala.get("i1")
    return (i1 - i0)*lambda_ + i0
```

```python
impacto_segmento(0.269671, "infra", escala_impacto)
```

```python
impacto_segmento(0.000221, "infra", escala_impacto)
```

```python
dcalc["impacto_segmento"] = dcalc.apply(lambda x : impacto_segmento(x["lambda"], x["segmento"], escala_impacto), axis=1)
```

#### Analises dos Resultados

```python
dcalc.drop(columns=["ocorrencias"], axis=1, inplace=True)
```

```python
color_menu = {"credito" : 'rgb(252,110,110)',
             "processos" : "rgb(146,233,249)",
              "infra" : "rgb(185,232,119)",
              "outros" : "rgb(247,241,86)"
             }

def plot_dividas(cnpj, df_dividas):
    dt = df_dividas[df_dividas["cnpj"]==cnpj]
    labels = dt["segmento"].tolist()
    values = dt["composicao"].tolist()
    values = [np.around(el*100, 2) for el in values]
    colors = [color_menu.get(el) for el in labels]
    data = []
    trace1 = go.Pie(
            labels = labels,
            values = values,
            hoverinfo = "label+percent",
            textinfo = "value",
            marker = dict(colors=colors, line = dict(color="rgb(0, 0, 0)", width=1))
    )
    
    layout = go.Layout(title="Composição da Dívida e Probabilidades: {}".format(cnpj),
                      
                      
                      )
    fig = go.Figure(data=[trace1], layout=layout)
    iplot(fig)
    return

def plot_probabilidade(cnpj, df_dividas):
    dt = df_dividas[df_dividas["cnpj"]==cnpj]
    labels = dt["segmento"].tolist()
    values = dt["probabilidade"].tolist()
    values = [np.around(el, 2) for el in values]
    colors = [color_menu.get(el) for el in labels]
    
    trace1 = go.Pie(
            labels = labels,
            values = values,
            hoverinfo = "label+percent",
            textinfo = "value",
            marker = dict(colors=colors, line = dict(color="rgb(0, 0, 0)", width=1))
    )
    
    
    
    layout = go.Layout(title="Probabilidades: {}".format(cnpj))
    fig = go.Figure(data=[trace1], layout=layout)
    iplot(fig)
    return

def plot_risco(cnpj, resp):
    dt = resp[resp['cnpj']==cnpj]
    score = int(dt["score"].iloc[0])
    dt.sort_values("risco", ascending=False, inplace=True)
    labels = dt["segmento"].tolist()
    data = []
    for el in labels:
        d = dt[dt['segmento']==el]
        trace = go.Bar(
            x = [el],
            y = [d["risco"].iloc[0]],
            marker = dict(color = color_menu.get(el), line=dict(color='rgb(0, 0, 0)', width=1)),
            name = el
        )
        data.append(trace)
    layout = go.Layout(title = "Risco por segmento de dívida - Score : {}".format(score), bargap=0.5
                      )
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
    return
```
