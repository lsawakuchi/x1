```python
def probabilidade(dt):
    dprob = dt.groupby("segmento").count().reset_index()[["segmento", "cnpj"]]
    dprob.columns = ["segmento", "ocorrencias"]
    dprob["probabilidade"] = dprob["ocorrencias"]/dprob["ocorrencias"].sum()
    return dprob

def composicao_dividas(dt):
    dimp = dt.groupby(["segmento"]).sum().reset_index()[["segmento", "valor"]]
    dimp["composicao"] = dimp["valor"]/dimp["valor"].sum()
    return dimp

escala_impacto = {"outros" : {"c0" : 0, "c1" : 0.25},
                 "infra" : {"c0" : 0.25, "c1" : 0.5},
                 "processos" : {"c0" : 0.5 , "c1" : 0.75},
                 "credito" : {"c0" : 0.75, "c1" : 1}
                 }

def get_impacto_segmento(comp, segmento, escala):
    escala_seg = escala.get(segmento)
    c0 = escala_seg.get("c0")
    c1 = escala_seg.get("c1")
    return (c1-c0)*comp + c0


def calcula_risco(cnpj, df_divida, escala_impacto):
    dt = df_divida[df_divida["cnpj"]==cnpj]
    dprob = probabilidade(dt)
    dimp = composicao_dividas(dt)
    dimp["impacto"] = dimp.apply(lambda x : get_impacto_segmento(x["composicao"], x["segmento"], escala_impacto), axis=1)
    drisco = dprob.merge(dimp, left_on="segmento", right_on="segmento", how="left")
    drisco["risco"] = drisco["impacto"]*drisco["probabilidade"]
    drisco["cnpj"] = cnpj
    return drisco

# score em funcao do risco

escala_score = {"credito" : {"s0" : 0, "s1" : 200},
               "processos" : {"s0" : 200, "s1" : 400},
                "infra" : {"s0" : 400, "s1" : 600},
                "outros" : {"s0" : 600, "s1" : 800}
               }

def calcula_dscore(r, seg, escala_score):
    escala = escala_score.get(seg)
    s0 = escala.get("s0")
    s1 = escala.get("s1")
    return (s0 - s1)*r + s0

```

```python
color_menu = {"credito" : 'rgb(252,110,110)',
             "processos" : "rgb(146,233,249)",
              "infra" : "rgb(185,232,119)",
              "outros" : "rgb(247,241,86)"
             }

def plot_dividas(cnpj, df_dividas):
    dt = df_dividas[df_dividas["cnpj"]==cnpj]
    labels = dt["segmento"].tolist()
    values = dt["composicao"].tolist()
    values = [np.around(el*100, 2) for el in values]
    colors = [color_menu.get(el) for el in labels]
    data = []
    trace1 = go.Pie(
            labels = labels,
            values = values,
            hoverinfo = "label+percent",
            textinfo = "value",
            marker = dict(colors=colors, line = dict(color="rgb(0, 0, 0)", width=1))
    )
    
    values = dt["probabilidade"].tolist()
    values = [np.around(el, 2) for el in values]
    trace2 = go.Pie(
            labels = labels,
            values = values,
            hoverinfo = "label+percent",
            textinfo = "value",
            marker = dict(colors=colors, line = dict(color="rgb(0, 0, 0)", width=1))
    )
    
    layout = go.Layout(title="Composição da Dívida e Probabilidades: {}".format(cnpj),
                      
                      
                      )
    fig = go.Figure(data=[trace1], layout=layout)
    iplot(fig)
    return

def plot_probabilidade(cnpj, df_dividas):
    dt = df_dividas[df_dividas["cnpj"]==cnpj]
    labels = dt["segmento"].tolist()
    values = dt["probabilidade"].tolist()
    values = [np.around(el, 2) for el in values]
    colors = [color_menu.get(el) for el in labels]
    
    trace1 = go.Pie(
            labels = labels,
            values = values,
            hoverinfo = "label+percent",
            textinfo = "value",
            marker = dict(colors=colors, line = dict(color="rgb(0, 0, 0)", width=1))
    )
    
    
    
    layout = go.Layout(title="Probabilidades: {}".format(cnpj))
    fig = go.Figure(data=[trace1], layout=layout)
    iplot(fig)
    return

def plot_risco(cnpj, resp):
    dt = resp[resp['cnpj']==cnpj]
    score = int(dt["score_final"].iloc[0])
    dt.sort_values("risco", ascending=False, inplace=True)
    labels = dt["segmento"].tolist()
    data = []
    for el in labels:
        d = dt[dt['segmento']==el]
        trace = go.Bar(
            x = [el],
            y = [d["risco"].iloc[0]],
            marker = dict(color = color_menu.get(el), line=dict(color='rgb(0, 0, 0)', width=1)),
            name = el
        )
        data.append(trace)
    layout = go.Layout(title = "Risco por segmento de dívida - Score : {}".format(score), bargap=0.5
                      )
    fig = go.Figure(data=data, layout=layout)
    iplot(fig)
    return
```

```python
from sqlalchemy import create_engine
from pricing.service.scoring.lscore import LScoring
```

```python
def get_escala_score(lscore):
    upper_limit = 0.8*lscore
    band = np.floor(upper_limit/4)
    escala = {"credito" : {"s0" : 0, "s1" : band},
               "processos" : {"s0" : band, "s1" : 2*band},
                "infra" : {"s0" : 2*band, "s1" : 3*band},
                "outros" : {"s0" : 3*band, "s1" : 4*band}
               }
    return escala
```
